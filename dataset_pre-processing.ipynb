#### Dataset split code for Deep Learning methods with split ratio: train:test = 85:15. 15% of the dataset (331 images) is approximately 50 images. To preserve the distribution across collection environments (Ground vs. Flight) in the test set, we selected 20 Ground and 30 Flight images

```python
import os
import shutil
import random

# === CONFIG ===
image_src = "../Automated-MIC-Detection/MIC_NASA_dataset/images"  
mask_src  = "../Automated-MIC-Detection/MIC_NASA_dataset/masks"  

# Output folders
train_img_dir = os.path.join(image_src, "..", "train", "images")
train_mask_dir = os.path.join(mask_src,  "..", "train", "masks")
test_img_dir  = os.path.join(image_src, "..", "test",  "images")
test_mask_dir = os.path.join(mask_src,  "..", "test",  "masks")

# === CLEAN DEST FOLDERS ===
for folder in [train_img_dir, train_mask_dir, test_img_dir, test_mask_dir]:
    if os.path.exists(folder):
        shutil.rmtree(folder)
    os.makedirs(folder, exist_ok=True)

# === LOAD FILENAMES ===
all_filenames = sorted([f for f in os.listdir(image_src) if f.lower().endswith(".png")])

# Case-insensitive prefix checks
flight_files = [f for f in all_filenames if f.lower().startswith("flight")]
ground_files = [f for f in all_filenames if f.lower().startswith("ground")]

# === SHUFFLE (reproducible) ===
random.seed(42)
random.shuffle(flight_files)
random.shuffle(ground_files)

# === EXACT TEST COUNTS ===
TEST_GROUND = 20
TEST_FLIGHT = 30

if len(ground_files) < TEST_GROUND:
    raise ValueError(f"Not enough Ground images: need {TEST_GROUND}, found {len(ground_files)}.")
if len(flight_files) < TEST_FLIGHT:
    raise ValueError(f"Not enough Flight images: need {TEST_FLIGHT}, found {len(flight_files)}.")

ground_test  = ground_files[:TEST_GROUND]
ground_train = ground_files[TEST_GROUND:]

flight_test  = flight_files[:TEST_FLIGHT]
flight_train = flight_files[TEST_FLIGHT:]

# Final splits
test_files  = ground_test + flight_test
train_files = ground_train + flight_train

# Safety checks
assert set(train_files).isdisjoint(set(test_files)), "Train/Test overlap detected."

# Ensure masks exist before moving anything
missing_masks = [f for f in (train_files + test_files) if not os.path.exists(os.path.join(mask_src, f))]
if missing_masks:
    raise FileNotFoundError(f"Missing mask files for: {missing_masks[:5]}{' ...' if len(missing_masks)>5 else ''}")

# === MOVE FILES ===
def move_pairs(filenames, img_dst, mask_dst):
    for fname in filenames:
        img_src_path  = os.path.join(image_src, fname)
        mask_src_path = os.path.join(mask_src,  fname)

        img_dst_path  = os.path.join(img_dst,  fname)
        mask_dst_path = os.path.join(mask_dst, fname)

        shutil.move(img_src_path,  img_dst_path)
        shutil.move(mask_src_path, mask_dst_path)

move_pairs(train_files, train_img_dir, train_mask_dir)
move_pairs(test_files,  test_img_dir,  test_mask_dir)

# === REPORT ===
print("Dataset split complete.")
print(f"Total files: {len(all_filenames)}")
print(f"Ground: {len(ground_files)} → Train: {len(ground_train)}, Test: {len(ground_test)}")
print(f"Flight: {len(flight_files)} → Train: {len(flight_train)}, Test: {len(flight_test)}")
print()
print(f"Moved to train/images: {len(os.listdir(train_img_dir))}")
print(f"Moved to train/masks : {len(os.listdir(train_mask_dir))}")
print(f"Moved to test/images : {len(os.listdir(test_img_dir))}")
print(f"Moved to test/masks  : {len(os.listdir(test_mask_dir))}")
```
